```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
span= 24.0
alpha = 2.0/(span+1)
```

```python
model = SimpleExpSmoothing(df['Load'])
```

```python
fitted_model = model.fit(smoothing_level=alpha,optimized=False)
```

```python
df['SES'] = fitted_model.fittedvalues.shift(-1)
```

```python
df.head()
```

```python
df[['Load','SES','1-hour-EWMA-Load']].iloc[0:1000].plot(figsize=(12,10))
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
df['DES_add'] = ExponentialSmoothing(df['Load'],trend='add').fit().fittedvalues.shift(-1)
```

```python
df[['Load','SES', 'DES_add']].iloc[0:500].plot(figsize=(15,15))
```

```python
df['DES_mul'] = ExponentialSmoothing(df['Load'],trend='mul').fit().fittedvalues.shift(-1)
```

```python
df[['Load','SES', 'DES_add','DES_mul']].iloc[0:100].plot(figsize=(15,15))
```
